# **Run the API Services**

```python
import os
import subprocess
import threading
from pyngrok import ngrok

# Change to the Django project directory
os.chdir('/content/drive/MyDrive/fish_snapid_svc')
print(os.getcwd())

def start_django_server():
    print("Start Django server...")

    # Start Django server and capture output
    process = subprocess.Popen(['python', 'manage.py', 'runserver', '0.0.0.0:8000'],
                               stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

    def read_output(pipe):
        while True:
            line = pipe.readline()
            if line:
                print(line, end='')
            else:
                break

    # Create threads to read stdout and stderr
    threading.Thread(target=read_output, args=(process.stdout,), daemon=True).start()
    threading.Thread(target=read_output, args=(process.stderr,), daemon=True).start()

start_django_server()

# Start ngrok tunnel
public_url = ngrok.connect(8000)
print(f"Public URL: {public_url}")

# Keep the ngrok tunnel open
!sleep infinity
```
